```python
import pandas as pd

# Ensure 'transaction_date' is of datetime type and remove rows with NaT
# This step is repeated here for robustness, in case the previous cell wasn't run immediately before
df_full['transaction_date'] = pd.to_datetime(df_full['transaction_date'], errors='coerce')
df_full.dropna(subset=['transaction_date'], inplace=True)

# Group by pdv, internal_product_id, and transaction_date (grouped by week), summing the quantity
df_weekly = df_full.groupby(
    ['pdv', 'internal_product_id', pd.Grouper(key='transaction_date', freq='W')]
)['quantity'].sum().reset_index()

print("Dados agregados por semana:")
display(df_weekly.head())
```

```python
# Ordenar os dados para garantir a consistência temporal
df_weekly = df_weekly.sort_values(by=['pdv', 'internal_product_id', 'transaction_date']).reset_index(drop=True)

# 1. Features de Tempo
df_weekly['semana_do_ano'] = df_weekly['transaction_date'].dt.isocalendar().week
df_weekly['mes'] = df_weekly['transaction_date'].dt.month
df_weekly['ano'] = df_weekly['transaction_date'].dt.year

# 2. Features de Lag (Vendas da semana anterior)
# O lag=1 significa que estamos pegando o valor da linha anterior
df_weekly['vendas_semana_anterior'] = df_weekly.groupby(['pdv', 'internal_product_id'])['quantity'].shift(1)

# 3. Features de Janela Móvel (Média das últimas 4 semanas)
# shift(1) para não usar a informação da própria semana no cálculo (evitar data leakage)
df_weekly['media_movel_4_semanas'] = df_weekly.groupby(['pdv', 'internal_product_id'])['quantity'].shift(1).rolling(window=4, min_periods=1).mean()

# Remover linhas com valores nulos criados pelos lags/rolling
df_weekly.dropna(inplace=True)

display(df_weekly.head(10))
```

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Definir data de corte (ex: últimas 5 semanas para validação)
data_corte = df_weekly['transaction_date'].max() - pd.DateOffset(weeks=5)
df_train = df_weekly[df_weekly['transaction_date'] <= data_corte]
df_val = df_weekly[df_weekly['transaction_date'] > data_corte]

# Definir features (X) e alvo (y)
features = ['semana_do_ano', 'mes', 'ano', 'vendas_semana_anterior', 'media_movel_4_semanas'] # Adicionar mais features que você criou
target = 'quantity'

X_train = df_train[features]
y_train = df_train[target]
X_val = df_val[features]
y_val = df_val[target]

print(f"Tamanho do treino: {len(X_train)}, Tamanho da validação: {len(X_val)}")
```

```python
# Instanciar e treinar o modelo XGBoost Regressor
model = xgb.XGBRegressor(
    objective='reg:squarederror',
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    n_jobs=-1, # Usar todos os cores da CPU
    early_stopping_rounds=50 # Parar de treinar se a performance não melhorar em 50 rodadas
)

model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    verbose=100 # Imprimir o progresso a cada 100 árvores
)
```

```python
# Supondo que você separou seus dados e tem um conjunto de teste 'X_test'
# Se você não tiver um X_test, pode usar o X_val para ver o desempenho na validação

# Fazer as previsões no conjunto de validação (X_val)
predictions = model.predict(X_val)

# As previsões serão armazenadas na variável 'predictions' como um array NumPy
# Vamos visualizar as 5 primeiras previsões geradas
print("As 5 primeiras previsões na validação são:")
print(predictions[:5])
```

```python
# É uma boa prática criar uma função para isso
def criar_features(df):
    """
    Cria novas features a partir de um DataFrame de vendas.
    """
    # Certifique-se que a coluna de data está no formato datetime
    df['transaction_date'] = pd.to_datetime(df['transaction_date'])

    # --- 1. Features de Tempo ---
    df['dia_da_semana'] = df['transaction_date'].dt.dayofweek  # Segunda=0, Domingo=6
    df['semana_do_ano'] = df['transaction_date'].dt.isocalendar().week.astype(int)
    df['mes'] = df['transaction_date'].dt.month
    df['dia_do_mes'] = df['transaction_date'].dt.day
    df['ano'] = df['transaction_date'].dt.year

    # --- 2. Features de Lag (ex: vendas da semana anterior) ---
    # ESSENCIAL: Os dados precisam estar ordenados por data para o lag funcionar!
    df = df.sort_values(by=['pdv', 'internal_product_id', 'transaction_date'])

    # O .shift(1) pega o valor da linha anterior
    # O groupby garante que estamos pegando o valor do mesmo produto/pdv
    df['venda_semana_anterior'] = df.groupby(['pdv', 'internal_product_id'])['quantity'].shift(1)

    # Preencher valores nulos (ex: primeira semana de um produto) com 0
    df['venda_semana_anterior'] = df['venda_semana_anterior'].fillna(0)


    # --- 3. Features de Eventos (Exemplo Hipotético) ---
    # Suponha que você descobriu que os erros acontecem em dias de promoção
    # Você precisaria criar essa lista com base no seu conhecimento de negócio
    datas_de_promocao = ['2024-11-29', '2024-12-24'] # Exemplo: Black Friday e Véspera de Natal
    datas_de_promocao = pd.to_datetime(datas_de_promocao)
    df['eh_promocao'] = df['transaction_date'].isin(datas_de_promocao).astype(int)

    return df

# Aplicar a função ao seu DataFrame completo
print("Criando novas features...")
df_com_features = criar_features(df_full.copy()) # Usar .copy() para evitar avisos
print("Novas features criadas!")
display(df_com_features.head())
```

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# --- Pré-requisitos ---
# Certifique-se de que 'df_com_features' seja o seu DataFrame
# com as colunas de tempo e de lag que criamos anteriormente.

# --- 1. PREPARAR OS DADOS (A Forma Correta) ---

print("1. Preparando e dividindo os dados cronologicamente...")

# Ordenar o DataFrame inteiro por data para garantir a sequência temporal
df_com_features['transaction_date'] = pd.to_datetime(df_com_features['transaction_date'])
df_com_features = df_com_features.sort_values(by='transaction_date').reset_index(drop=True)

# Definir as features (X) e o alvo (y)
features = [col for col in df_com_features.columns if col not in ['transaction_date', 'quantity'] and df_com_features[col].dtype != 'object']
X = df_com_features[features]
y = df_com_features['quantity']

# Calcular o ponto de corte para a divisão (80% para treino, 20% para teste)
ponto_de_corte = int(len(df_com_features) * 0.8)

# Dividir os dados em passado (treino) e futuro (teste)
X_train = X.iloc[:ponto_de_corte]
y_train = y.iloc[:ponto_de_corte]

X_test = X.iloc[ponto_de_corte:]
y_test = y.iloc[ponto_de_corte:]

print("Dados divididos com sucesso!")
print(f"Treino: {len(X_train)} amostras | Teste: {len(X_test)} amostras")


# --- 2. CRIAR E TREINAR O MODELO ---

print("\n2. Criando e treinando o modelo XGBoost...")

# Instanciar o modelo com os mesmos parâmetros de antes
model = xgb.XGBRegressor(
    objective='reg:squarederror',
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    n_jobs=-1,  # Usar todos os processadores
    random_state=42 # Para reprodutibilidade
)

# Treinar o modelo usando os dados de treino cronológicos
# Adicionamos um eval_set para monitorar o desempenho, o que é uma boa prática
model.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    verbose=False # Mude para True ou um número (ex: 100) se quiser ver o progresso
)

print("Modelo treinado com sucesso!")


# --- 3. FAZER PREVISÕES E AVALIAR ---

print("\n3. Fazendo previsões e avaliando o desempenho...")

# Fazer previsões no conjunto de teste (os dados do "futuro")
predictions = model.predict(X_test)

# Calcular as métricas de desempenho
r2 = r2_score(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))

# --- IMPRIMIR O RESULTADO FINAL ---
print("\n--- RESULTADOS FINALS DO MODELO ---")
print(f"Coeficiente de Determinação (R²): {r2:.4f}")
print(f"Erro Absoluto Médio (MAE): {mae:.4f}")
print(f"Raiz do Erro Quadrático Médio (RMSE): {rmse:.4f}")
```
